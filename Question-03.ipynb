```python
class Person:
    def __init__(self):
        '''Person class constructor, with no parameters to be passed'''
        self.requested_floor = ""
        self.person_weight = 0
    def set_requested_floor(self,floor_number):
        '''sets the requested floor for the person'''
        self.requested_floor = floor_number
    def get_requested_floor(self):
        '''returns the requested floor by person'''
        return self.requested_floor
    def set_person_weight(self,weight):
        '''sets the person weight who requested the floor'''
        self.person_weight = weight
    def get_person_weight(self):
        '''returns the person weight who requested the floor'''
        return self.person_weight
        
        
```

```python
class Elevator:
    def __init__(self,floors):
        '''Constructor for Elevator called, with number of floors to be
        passed as parameter'''
        self.current_floor = ""               #saves current floor
        self.current_weight = 0               #saves current weight of lift
        self.requested_floors = []            #list of requested floors by persons
        self.weight_capacity = 200            #setting the  lift weight capacity
        self.persons_list = []                #stores the list of persons who entered in lift
        self.number_of_floors = floors        #stores number of floors which lift need to be travelled
    def capacity_satisfied(self):
        '''Checks and returns bool by checking if capacity of lift
        is statisfied by requested weight'''
        if self.current_weight<=self.weight_capacity:
            return True
        else:
            return False
    def enter_person(self,a_person):
        '''Enters the person into lift, by checking the lift weight
        capacity and also if the requested floor by person exists
        or not'''
        if self.capacity_satisfied():
            if a_person.get_requested_floor()<=self.number_of_floors:
                #adding person to the persons list
                self.persons_list.append(a_person)
                self.requested_floors.append(a_person.get_requested_floor())
                self.current_weight += + a_person.get_person_weight()
            else:
                print("Sorry, request floor by person # {} does not exists!".format(len(self.persons_list)+1))
        else:
            print("Sorry! lift capacity reached!")
            print("Person # {} cannot be entered!".format(len(self.persons_list)+1))
    def open_door(self):
        '''opens the door of lift'''
        print("Opening the door...")
    def close_door(self):
        '''closes the door of lift'''
        print("Closing the door")
    def go_up(self):
        '''Lift goes up'''
        print("Lift going up...")
    def go_down(self):
        '''lift goes down'''
        print("Lift going down...")
    def run_elevator(self):
        '''This function runs the elevator by picking up requested floors
        with optimization and then announcing the particular floor reached'''
        self.requested_floors.sort()
        for floor_number in self.requested_floors:
            print("Floor # {} reached!".format(floor_number))
            self.open_door()
            self.close_door()
            print()
        
    
    
        
```
