```python
from datetime import datetime,date
class CellPhone:
    def __init__(self,color,model,storage,carrier):
        '''Constructor for CellPhone class with parameters:
        color, model, storage, carrier'''
        self.color = color
        self.model = model
        self.storage = storage
        self.carrier = carrier
        self.old_storage=0 
    def set_color(self,color):
        '''sets the color of mobile'''
        self.color = color
    def set_model(self,model):
        '''sets the model of phone'''
        self.model=model
    def set_storage(self,storage):
        '''sets the storage of phone'''
        self.storage=storage
    def set_carrier(self,carrier):
        '''sets the carrier of phone'''
        self.carrier=carrier
    def get_color(self):
        '''returns the color of phone'''
        return "Color is {}".format(self.color)
    def get_model(self):
        '''returns the model of phone'''
        return "Model is {}".format(self.model)
    def get_storage(self):
        '''returns the storage of phone'''
        return self.storage
#         return "Storage is {} GB".format(self.storage)
    def get_carrier(self):
        '''returns the carrier of phone'''
        return "Carrier is {}".format(self.carrier)
    def __str__(self):
        '''returns the cell phone information as a string'''
        return "This is {} phone with model {} and storage of {}, having carrier {}".format(self.color,self.model,self.storage,self.carrier)
    def upgrade_storage(self,storage):
        '''upgrades the storage of phone'''
        print("Old storage is {}".format(self.storage))
        self.storage = storage
        print("Updated storage is {}".format(self.storage))
    def connect_cloud(self):
        '''connects phone to cloud'''
        print("connecting to cloud...")
    def download_app(self):
        '''downloads application from playstores'''
        print("Downloading app from playstore...")
    def get_time(self):
        '''returns the current time'''
        from datetime import datetime
        now = datetime.now()
        current_time = now.strftime("%I:%M %p")
        print("Time is:", current_time)
    def get_date(self):
        '''returns the current date'''
        today = date.today()
        print("Today's date:", today)
    def place_call(self,number):
        '''Makes call to  given number'''
        print("Calling number {}...".format(number))
    def send_sms(self,msg,number):
        '''sends given message to given number'''
        print("Sending Message '{}' to number {}".format(msg,number))
        
        
```

```python
iPhone = ApplePhone('Black', '11 Pro', 64, 'Verizon')    #making object of Apple Phone
iPhone.set_airdrop_status('On');                         #sets the air drop status on
print(iPhone)                                            #prints the Apple Phone
iPhone.get_storage()                                     #gets the storage of Apple Phone
iPhone.upgrade_storage(128)                              #upgrades the storage of Apple Phone
iPhone.connect_cloud()                                   #connects to icloud
iPhone.download_app('Starbucks')                         #downloads given application from apple store
iPhone.get_time()                                        #gets the current time
iPhone.get_date()                                        #gets the current date
iPhone.place_call('425-123-4567')                        #makes call to given number
iPhone.send_sms('Hi There', '214-214-2144')              #sends given message to given number
iPhone.launch_Siri()                                     #launches siri
print('AirDrop Status is: ', iPhone.get_airdrop_status())#gets apple phone airdrop status
print("Storage is {} GB".format(iPhone.get_storage()))
```
