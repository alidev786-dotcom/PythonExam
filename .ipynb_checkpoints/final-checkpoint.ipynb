```python
from datetime import datetime,date
class CellPhone:
    def __init__(self,color,model,storage,carrier):
        '''Constructor for CellPhone class with parameters:
        color, model, storage, carrier'''
        self.color = color
        self.model = model
        self.storage = storage
        self.carrier = carrier
        self.old_storage=0 
    def set_color(self,color):
        '''sets the color of mobile'''
        self.color = color
    def set_model(self,model):
        '''sets the model of phone'''
        self.model=model
    def set_storage(self,storage):
        '''sets the storage of phone'''
        self.storage=storage
    def set_carrier(self,carrier):
        '''sets the carrier of phone'''
        self.carrier=carrier
    def get_color(self):
        '''returns the color of phone'''
        return "Color is {}".format(self.color)
    def get_model(self):
        '''returns the model of phone'''
        return "Model is {}".format(self.model)
    def get_storage(self):
        '''returns the storage of phone'''
        return self.storage
#         return "Storage is {} GB".format(self.storage)
    def get_carrier(self):
        '''returns the carrier of phone'''
        return "Carrier is {}".format(self.carrier)
    def __str__(self):
        '''returns the cell phone information as a string'''
        return "This is {} phone with model {} and storage of {}, having carrier {}".format(self.color,self.model,self.storage,self.carrier)
    def upgrade_storage(self,storage):
        '''upgrades the storage of phone'''
        print("Old storage is {}".format(self.storage))
        self.storage = storage
        print("Updated storage is {}".format(self.storage))
    def connect_cloud(self):
        '''connects phone to cloud'''
        print("connecting to cloud...")
    def download_app(self):
        '''downloads application from playstores'''
        print("Downloading app from playstore...")
    def get_time(self):
        '''returns the current time'''
        from datetime import datetime
        now = datetime.now()
        current_time = now.strftime("%I:%M %p")
        print("Time is:", current_time)
    def get_date(self):
        '''returns the current date'''
        today = date.today()
        print("Today's date:", today)
    def place_call(self,number):
        '''Makes call to  given number'''
        print("Calling number {}...".format(number))
    def send_sms(self,msg,number):
        '''sends given message to given number'''
        print("Sending Message '{}' to number {}".format(msg,number))
        
        
```

```python
class Calculator:
    def __init__(self,first=0,second=0):
        '''Constructor method for Calculator class with paramters:
        first operand, second operand'''
        self.first_num = first
        self.second_num = second
        self.result = 0
        self.power_mode = ""
    def set_power_mode(self,mode):
        '''Sets the mode of power of calculator'''
        self.power_mode = mode
    def get_power_mode(self):
        '''returns the mode of power of calculator'''
        return self.power_mode
    def set_first_num(self,num1):
        '''sets the first operand of calculator'''
        self.first_num=num1
    def set_second_num(self,num2):
        '''sets the second operand of calculator'''
        self.second_num=num2
    def set_operator(self,opt):
        '''sets the operator which need to be performed'''
        self.operator=opt
    def get_first_num(self):
        '''returns the first operand of calculator'''
        return self.first_num
    def get_second_num(self):
        '''returns the second operand of calculator'''
        return self.second_num
    def add(self):
        '''Function to add 2 numbers'''
        self.result = self.first_num + self.second_num
    def subtract(self):
        '''Function to subtract 2 numbers'''
        self.result = self.first_num - self.second_num
    def multiply(self):
        '''Function to multiply 2 numbers'''
        self.result = self.first_num * self.second_num
    def divide(self):
        '''Function to divide 2 numbers'''
        self.result = self.first_num / self.second_num
    def get_result(self):
        '''returns the result of operation'''
        return self.result
```

```python
myCalc = Calculator(10,10)         #making calculator object with n1=10,n2=10
myCalc.set_power_mode("solar")     #sets the power mode of calculator
myCalc.add()                       #adds two numbers
print("{} + {} : {}".format(myCalc.get_first_num(),myCalc.get_second_num(),myCalc.get_result()))
myCalc.multiply()                  #multiply two numbers
print("{} x {} : {}".format(myCalc.get_first_num(),myCalc.get_second_num(),myCalc.get_result()))
print("Power mode: {}".format(myCalc.get_power_mode()))  #displays power mode
```

```python
import math
class AdvancedCalculator(Calculator):
    def __init__(self,first=0,second=0):
        '''Constructor for AdvancedCalculator class with parameters:first operand, second operand'''
        super().__init__(first,second)
    def sqrt(self):
        '''calculates square root of first operand '''
        self.result = math.sqrt(self.first_num)
    def power(self):
        ''' calculates power of first operand to the exponent second
        operand'''
        self.result = math.pow(self.first_num,self.second_num)
    def logarithm(self):
        ''' calculates the logarithm of first operand to the base of
        second operand'''
        self.result = math.log(self.first_num,self.second_num)
    def sin(self):
        ''' calculates the sin of first operand'''
        self.result = math.sin(self.first_num)
    def cos(self):
        ''' calculates the cos of first operand'''
        self.result = math.cos(self.first_num)
    def tan(self):
        ''' calculates the tan of first operand'''
        self.result = math.tan(self.first_num)
```

```python
cal = AdvancedCalculator(10,4)
cal.sqrt()
print("Sqrt {}: {}".format(cal.get_first_num(),cal.get_result()))
cal.power()
print("{} ^ {} : {} ".format(cal.get_first_num(),cal.get_second_num(),cal.get_result()))
cal.logarithm()
print("log of {} to the base {} : {}".format(cal.get_first_num(),cal.get_second_num(),cal.get_result()))
cal.tan()
print("tan of {} : {}".format(cal.get_first_num(),cal.get_result()))
```

```python
class Person:
    def __init__(self):
        '''Person class constructor, with no parameters to be passed'''
        self.requested_floor = ""
        self.person_weight = 0
    def set_requested_floor(self,floor_number):
        '''sets the requested floor for the person'''
        self.requested_floor = floor_number
    def get_requested_floor(self):
        '''returns the requested floor by person'''
        return self.requested_floor
    def set_person_weight(self,weight):
        '''sets the person weight who requested the floor'''
        self.person_weight = weight
    def get_person_weight(self):
        '''returns the person weight who requested the floor'''
        return self.person_weight
```

```python
class Elevator:
    def __init__(self,floors):
        '''Constructor for Elevator called, with number of floors to be
        passed as parameter'''
        self.current_floor = ""               #saves current floor
        self.current_weight = 0               #saves current weight of lift
        self.requested_floors = []            #list of requested floors by persons
        self.weight_capacity = 200            #setting the  lift weight capacity
        self.persons_list = []                #stores the list of persons who entered in lift
        self.number_of_floors = floors        #stores number of floors which lift need to be travelled
    def capacity_satisfied(self):
        '''Checks and returns bool by checking if capacity of lift
        is statisfied by requested weight'''
        if self.current_weight<=self.weight_capacity:
            return True
        else:
            return False
    def enter_person(self,a_person):
        '''Enters the person into lift, by checking the lift weight
        capacity and also if the requested floor by person exists
        or not'''
        if self.capacity_satisfied():
            if a_person.get_requested_floor()<=self.number_of_floors:
                #adding person to the persons list
                self.persons_list.append(a_person)
                self.requested_floors.append(a_person.get_requested_floor())
                self.current_weight += + a_person.get_person_weight()
            else:
                print("Sorry, request floor by person # {} does not exists!".format(len(self.persons_list)+1))
        else:
            print("Sorry! lift capacity reached!")
            print("Person # {} cannot be entered!".format(len(self.persons_list)+1))
    def open_door(self):
        '''opens the door of lift'''
        print("Opening the door...")
    def close_door(self):
        '''closes the door of lift'''
        print("Closing the door")
    def go_up(self):
        '''Lift goes up'''
        print("Lift going up...")
    def go_down(self):
        '''lift goes down'''
        print("Lift going down...")
    def run_elevator(self):
        '''This function runs the elevator by picking up requested floors
        with optimization and then announcing the particular floor reached'''
        self.requested_floors.sort()
        for floor_number in self.requested_floors:
            print("Floor # {} reached!".format(floor_number))
            self.open_door()
            self.close_door()
            print()
```
